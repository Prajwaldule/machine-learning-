```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
# Load tokenizer and model correctly
tokenizer = AutoTokenizer.from_pretrained("joeddav/distilbert-base-uncased-go-emotions-student")
model = AutoModelForSequenceClassification.from_pretrained("joeddav/distilbert-base-uncased-go-emotions-student")
```

```python
text = "Your dedication to your work is truly commendable."
```

```python
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)

# Get predicted sentiment label
labels = model.config.id2label
predicted_label_index = torch.argmax(outputs.logits).item()
sentiment = labels[predicted_label_index]

# Print the sentiment
print("Sentiment:", sentiment)
```
